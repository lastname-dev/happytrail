# 축제 URL 가져오기

```python
# 축제 정보를 저장할 url
festival_urls = []
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# ChromeDriver 경로 설정
driver_path = "./chromedriver.exe"

# WebDriver 초기화
driver = webdriver.Chrome(driver_path)

# 해당 사이트 불러오기
url = "https://event-us.kr/search/festival"
driver.get(url)

# 더보기 버튼 클릭 및 스크롤 내리기
while True:
    # 현재 페이지에서 모든 <a> 태그 가져오기
    links = driver.find_elements(By.TAG_NAME, "a")

    # 각 링크 저장
    for link in links:
        href = link.get_attribute('href')
        if href is not None and 'id=' in href:
            festival_urls.append(href)

    # 더보기 버튼 클릭
    try:
        more_button = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, ".flex.justify-center.items-center.px-4.w-full.rounded-md.border.bg-white.font-normal.shadow-xs.py-4.text-sm.border-none.text-gray-500.mt-20.leading-none.border-none.text-gray-500.mt-20")))
        driver.execute_script("arguments[0].scrollIntoView(true);", more_button)
        more_button.click()
        # 추가적인 콘텐츠가 로드될 때까지 대기
        time.sleep(1)  # 적절한 대기 시간 설정
    except:
        break

# 웹 드라이버 종료
driver.quit()
```

```python
festival_urls = set(festival_urls) # 중복 제거
festival_urls = list(festival_urls) # 다시 list로
print(len(festival_urls)) # 총 몇개의 축제인지 확인 
festival_urls
```

# 가져온 URL을 토대로 정보 추출하기

```python
title = [] # 축제 이름
urls = [] # 대표 이미지
```

```python
infos = []
```

```python
from tqdm import tqdm
from selenium.common.exceptions import TimeoutException

for url in tqdm(festival_urls):
    try :
        # ChromeDriver 경로 설정
        driver_path = "./chromedriver.exe"

        # WebDriver 초기화
        driver = webdriver.Chrome(driver_path)

        driver.get(url)

        # 축제 이름
        name = driver.find_elements(By.CSS_SELECTOR,".text-lg.font-bold")
        for element in name:
            e = element.text
            if(e=="" or e==None) : continue
            title.append(e)

        # 이미지
        img = driver.find_elements(By.CSS_SELECTOR,".w-full.rounded-md")
        for element in img:
            src = element.get_attribute('src')
            if src=="" or src==None : continue
        #     print(src)
            urls.append(src)

        # 주소 ~ 설명
        info = []
        tmp = driver.find_elements(By.CSS_SELECTOR, "dd.mt-1.lg\\:mt-0.lg\\:col-span-6")
        for i in range(1, 8):
            e = tmp[i].text
            if(e=="" or e==None or e[0]=='(') : continue
            info.append(e)
        infos.append(info)
        driver.quit()
        
    except TimeoutException:
        pass
```

```python
addrs = [] # 주소
places = [] # 장소
startdates = [] # 축제 시작일
enddates = [] # 축제 종료일
descs = [] # 설명

remove_idx = []
for i in tqdm(range(0, len(infos))):
    info = infos[i]
    if len(info) < 5:
        remove_idx.append(i)
        continue
    descs.append(info[0])
    addrs.append(info[1])
    places.append(info[2])
    startdates.append(info[3])
    enddates.append(info[4])
```

```python
import copy

# 깊은 복사
modified_title = copy.deepcopy(title)
modified_urls = copy.deepcopy(urls)

# 인덱스 제거
for index in sorted(remove_idx, reverse=True):
    del modified_title[index]
    del modified_urls[index]

# 원본 리스트 출력
# print(title)
# print(urls)

# 수정된 리스트 출력
print(len(modified_title))
print(len(modified_urls))
```

```python
df = pd.DataFrame({
    "title": modified_title,
    "url": modified_urls,
    "address": addrs,
    "place": places,
    "desc" : descs,
    "startdate" : startdates,
    "enddate" : enddates
})
```
